```python
import pandas as pd
import numpy as np
import datetime
```

```python
df_gg = pd.read_csv('/Users/lauralayton/Downloads/greenhouse_gas.csv')

# NDVI_2003_2012 = pd.read_csv('/Users/lauralayton/Downloads/Wetland_NDVI_2003_2012.csv')

NDVI_2003_MAR2013 = pd.read_csv('/Users/lauralayton/Downloads/Wetland_NDVI_2003_MAR2013.csv')
NDVI_2013_2016 = pd.read_csv('/Users/lauralayton/Downloads/Wetland_NDVI_2013_2016.csv')

# concatenate into one NDVI data frame
# NDVI_2003_2016 = pd.concat([NDVI_2003_2012, NDVI_2013_2016], axis = 0)
NDVI_2003_2016 = pd.concat([NDVI_2003_MAR2013, NDVI_2013_2016], axis = 0)

# filter df_gg for only those wetlands with data also in the NDVI spreadsheet
df = df_gg[(df_gg['WETLAND_ID'].isin(NDVI_2003_2016['WETLANDID'].unique()))]

# now filter NDVI_2003_2016 for only wetland IDs also in df
NDVI_2003_16 = NDVI_2003_2016[NDVI_2003_2016['WETLANDID'].isin(df['WETLAND_ID'].unique())]
```

```python
pd.options.mode.chained_assignment = None  # default='warn'

# constructing the various date columns for both dataframes

# breaking the date into day, month, and year for both dataframes

def convert_gg_time_data(date_time):
    format = '%d%b%Y'
    datetime_str = datetime.datetime.strptime(date_time, format)

    return datetime_str

df['Date Object'] = df['DATE'].apply(convert_gg_time_data)
df['Date Object'] = pd.to_datetime(df['Date Object'])
df['Day'] = pd.to_datetime(df['Date Object']).dt.day
df['Month'] = pd.to_datetime(df['Date Object']).dt.month
df['Year'] = pd.to_datetime(df['Date Object']).dt.year


def make_datestr_ndvi(timestr):
    datestr = timestr[0:10]
    return datestr


NDVI_2003_16['Date String'] = NDVI_2003_16['imageId'].apply(make_datestr_ndvi)
NDVI_2003_16['Date Object'] = pd.to_datetime(NDVI_2003_16['Date String'])
NDVI_2003_16['Day'] = pd.to_datetime(NDVI_2003_16['Date Object']).dt.day
NDVI_2003_16['Month'] = pd.to_datetime(NDVI_2003_16['Date Object']).dt.month
NDVI_2003_16['Year'] = pd.to_datetime(NDVI_2003_16['Date Object']).dt.year
```

```python
# now I suppose we toss out any years that don't match
# ok so spoilers, though df_gg has data from back to 2003, it only starts in 2009 for these wetlands apparently

NDVI_2009_16 = NDVI_2003_16[NDVI_2003_16['Year'].isin(df['Year'].unique())]
```

```python
# now testing to see how I can match up the dates and NDVI values

count_not_in_NDVI = 0
count_in_NDVI = 0

for index, row in df.iterrows():
    year = row['Year']
    month = row['Month']
    day = row['Day']

    NDVI_subset = NDVI_2009_16[(NDVI_2009_16['Year'] == year) &
        (NDVI_2009_16['Month'] == month) & 
        (NDVI_2009_16['Day'] == day)]

    if len(NDVI_subset) >= 1:
        # print(str(year) + " " + str(month) + " " + str(day))
        count_in_NDVI += 1
    else:
        count_not_in_NDVI += 1


print("in NDVI: " + str(count_in_NDVI))
print("not in NDVI: " + str(count_not_in_NDVI))
```

```python
# okay now I must attach the NDVI values to the correct wetland and dates

def select_closest_date(date1, dateslst, tolerance):
    min_diff = 365
    idx = 0

    for i in range(len(dateslst)):
        date = dateslst[i]
        diff = abs((date1 - date).days)
        if diff < min_diff:
            min_diff = diff
            idx = i

    if min_diff > tolerance:
        print("ERRORRRRR")
        print(date1)
        print(dateslst[idx])
        print(min_diff)
    return dateslst[idx]

    

df['NDVI'] = np.nan

for index, row in df.iterrows():
    wetland_ID = row['WETLAND_ID']
    date = row['Date Object']
    NDVI_wetland_subset = NDVI_2009_16[NDVI_2009_16['WETLANDID'] == wetland_ID]

    if len(NDVI_wetland_subset) < 1:
        print("***************HELPPPPP*****************")
        # ok this doesn't seem to occur

    date_to_use = select_closest_date(date, NDVI_wetland_subset['Date Object'].values, tolerance)

    df.loc[index, 'NDVI'] = NDVI_wetland_subset[NDVI_wetland_subset['Date Object'] == date_to_use]['mean'].values[0]
```
