```python
def water_ind(val):
    if val == 2:
        return 1
    else:
        return 0

def no_data_ind(val):
    if val == 0:
        return 1
    else:
        return 0

def indicators(df_gsw):
    df_gsw['NO_DATA_IND'] = df_gsw['water'].apply(no_data_ind)
    df_gsw['WATER_IND'] = df_gsw['water'].apply(water_ind)
```

```python
df_lst = [gsw_09,
          gsw_10,
          gsw_11,
          gsw_12,
          gsw_13,
          gsw_14,
          gsw_15,
          gsw_16]

for df in df_lst:
    indicators(df)
    df['HYDROPERIOD'] = np.nan
    df['MONTHS_NO_DATA'] = np.nan
    hloc = df.columns.get_loc('HYDROPERIOD')
    mloc = df.columns.get_loc('MONTHS_NO_DATA')
    
    for index, row in df.iterrows():
        lat = row['LAT']
        long = row['LONG']
        wID = row['Wetland_ID']
        
        df_subset = df[(df['LAT'] == lat) & 
            (df['LONG'] == long) & 
            (df['Wetland_ID'] == wID)]
    
        hydroperiod = 30.5 * sum(df_subset['WATER_IND'])
        mon_no_data = sum(df_subset['NO_DATA_IND'])
    
        df.iloc[index, hloc] = hydroperiod
        df.iloc[index, mloc] = mon_no_data
```

```python
# Now concatenate the dataframes together to get one df containing 2009-2016 data
gsw_09_16 = pd.concat(df_lst)

# Re-index the concatenated df
GSW_09_16 = gsw_09_16.reset_index(drop = True)
```

```python
# Make date functions to create date objects
def convert(date_str):
    format = '%Y-%m-%d'
    datetime_str = datetime.datetime.strptime(date_str, format)

    return datetime_str


# function to extract the date string from the imageId column in the ndvi dataset
def extract_date(imageId_str):
    return imageId_str[0:10]

# Re-index the concatenated df
GSW_09_16 = gsw_09_16.reset_index(drop = True)

# make the date columns in each dataframe
GSW_09_16['DATE'] = GSW_09_16['DATE'].apply(convert)
ndvi_03_12['DATE'] = ndvi_03_12['imageId'].apply(extract_date)
ndvi_13_16['DATE'] = ndvi_13_16['imageId'].apply(extract_date)
ndvi_03_12['DATE'] = ndvi_03_12['DATE'].apply(convert)
ndvi_13_16['DATE'] = ndvi_13_16['DATE'].apply(convert)

# now only extract NDVI values from 2009-2016
ndvi_09_12 = ndvi_03_12[(ndvi_03_12['DATE']).dt.year >= 2009]

# create a df containing NDVI from 2009 through 2016
ndvi_09_16 = pd.concat([ndvi_09_12, ndvi_13_16])

# Re-index the concatenated df
NDVI_09_16 = ndvi_09_16.reset_index(drop = True)
```

```python
def determine_nearest_date(date, datelst):
    ind_nearest = 0
    min_diff = (date - datelst[ind_nearest]).days

    for i in range(len(datelst)):
        dt = datelst[i]
        diff = (date - dt).days
        if abs(diff) < abs(min_diff):
            min_diff = diff
            ind_nearest = i

    return [datelst[ind_nearest], min_diff]
```

```python
# Add NDVI data to the GSW df based on date proximity
GSW_09_16['MEAN NDVI'] = np.nan

ndviloc = GSW_09_16.columns.get_loc('MEAN NDVI')

for index, row in GSW_09_16.iterrows():
    # extract the specific wetland and year that we want
    wID = row['Wetland_ID']
    date = row['DATE']
    year = date.year

    # subset the NDVI df for that wetland and year
    ndvi_subset = NDVI_09_16[(NDVI_09_16['WETLANDID'] == wID) & 
        (NDVI_09_16['DATE'].dt.year == year)]

    # determine the nearest NDVI observation date (relative to the date of the GSW observation)
    date_results = determine_nearest_date(date, ndvi_subset['DATE'].tolist())
    nearest_date = date_results[0]
    date_diff = date_results[1]
    # print(str(date_diff) + ", " + str(date) + ", " + str(nearest_date) + str(ndvi_subset['DATE'].tolist()))
    if abs(date_diff) <= 30:
        meanNDVI = ndvi_subset[ndvi_subset['DATE'] == nearest_date]['mean']
        GSW_09_16.iloc[index, ndviloc] = meanNDVI
    
```
